Outliers Detection


```python
import pandas as pd
```

```python
df=pd.read_csv("/height3.csv")
df
```

```python
df['height'].quantile(0.95)
```

```python
# Finding Percentile
max_threshold=df.height.quantile(0.95)
min_threshold=df.height.quantile(0.005)
max_threshold,min_threshold
```

```python
df[df['height']>max_threshold]
```

```python
df[df['height']<min_threshold]
```

```python
# Assingn the quantile by maximum threshold and minimum_threshold

df_no_outliers=df[(df['height']<max_threshold) & (df['height']> min_threshold) ]
df
```

```python
df_no_outliers=df[(df['height']>max_threshold) | (df['height']< min_threshold) ]
df_no_outliers
```
